3. Memeriksa apakah data tersebut memiliki null dan cara mengisi nilai null

```python
# Data studentVle
print(dataset_1.isnull().sum())
```

```python
# Data vle
print(dataset_2.isnull().sum())
```

```python
# Dikarenakan dalam data vle memiliki null value, maka harus di isi agar data tidak miss value
for col in dataset_2.columns:
    mode_value = dataset_2[col].mode()[0]
    dataset_2[col].fillna(mode_value, inplace=True)
print(dataset_2.isnull().sum())
```
